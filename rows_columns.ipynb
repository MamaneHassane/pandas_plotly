Ajoutons une colonne age

```python
def calculer_age(date_naissance):
    if isinstance(date_naissance, str):
        date_naissance = datetime.strptime(date_naissance, "%Y-%m-%d")
    aujourdhui = datetime.today()
    age = aujourdhui.year - date_naissance.year
    if (aujourdhui.month, aujourdhui.day) < (date_naissance.month, date_naissance.day):
        age -= 1
    return age

df['age'] = pd.to_datetime(df['date_of_birth'], dayfirst=True).apply(calculer_age)
df = df[sorted(df.columns)]
df
```
