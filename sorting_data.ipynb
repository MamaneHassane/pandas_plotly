```python
import pandas as pd
```

```python
data = pd.read_csv('data/data.csv')
df = pd.DataFrame(data)
df
```

Trier par colonnes et/ou ordre

```python
df.sort_values(by=['first_name', 'last_name'], ascending=True, inplace=True)
df
```

```python
df.sort_values(by=['first_name', 'last_name'], ascending=[True, False])
df
```

Remettre Ã  dÃ©faut

```python
df.sort_index(ascending=True, inplace=True)
df
```

Trier une colonne

```python
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'], dayfirst=True)
df['date_of_birth'].sort_values()
```

Une petite analyse des salaires

```python
largest_salaries = df['salary'].nlargest(100).index
df.loc[largest_salaries]
```

```python
salaire_par_domaines = df[['field', 'salary']]
salaire_par_domaines.sort_values(by=['field', 'salary'], ascending=[True, True])
```
